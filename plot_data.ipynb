# Plot data

## Import packages

```python
import matplotlib.pyplot as plot
import numpy as np
```

## Load data in CSV format

```python
data = np.loadtxt('./data/Virtual_TEMP.txt', delimiter=',', skiprows=1)
```

Check data size

```python
data.shape
```

## Plot

```python
fig, (a1, a2) = plot.subplots(2,1, sharex=True)
a1.set_ylabel('pwm')
a1.plot(data[:,0], data[:,2], 'g-')
a2.set_xlabel('time')
a2.set_ylabel('temperature')
a2.plot(data[:,0], data[:,1], 'r-')
```
